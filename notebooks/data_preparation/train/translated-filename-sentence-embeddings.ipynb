```python
import os

import numpy as np
import pandas as pd
import cloudpickle

from transformers import XLMRobertaModel, XLMRobertaTokenizer
import torch

from sentence_transformers import SentenceTransformer, util
```

```python
df = pd.read_csv('../input/final-image-traindataset-part0-4-count-5/final_images.csv', keep_default_na=False)

df
```

### Loading Model

```python
model = SentenceTransformer('paraphrase-multilingual-mpnet-base-v2')
```

## Retrieving Embeddings

```python
filename_embeddings = model.encode(df['PREfinal_filename'].values)

filename_embeddings.shape
```

```python
with open('translated_filename_embeddings.data', 'wb') as file:
    cloudpickle.dump(filename_embeddings, file)
```
