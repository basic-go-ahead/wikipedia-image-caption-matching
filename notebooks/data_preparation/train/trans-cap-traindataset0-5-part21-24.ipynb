```python
import numpy as np
import pandas as pd

from deep_translator import GoogleTranslator
```

```python
N_PARTS = 24
PART = 21 # in [0, 23]

field = 'undigit_caption'
field_lang = 'caption_lang'
field_translation = 'caption_translation'
```

```python
df = pd.read_csv('../input/traindataset-part0-4-count-5/matchings_part0_between4,5.csv')

df
```

```python
df[field_translation] = ''

df[field] = df[field].fillna('')

indices = df[(df[field_lang] != 'en') & (df[field] != '') & (~df[field].str.isspace())].index

print(len(indices))

df
```

```python
# N = 2112
# print(df.loc[N, 'caption_reference_description'])
# translator.translate_batch([df.loc[N, 'caption_reference_description']])
```

```python
SIZE = len(indices)

if PART == 0:
    r = range(SIZE // N_PARTS)
elif PART == N_PARTS - 1:
    r = range((N_PARTS - 1) * SIZE // N_PARTS, SIZE)
else:
    r = range((PART) * SIZE // N_PARTS, (PART + 1) * SIZE // N_PARTS)
    
print(min(r), max(r))
```

```python
translator = GoogleTranslator(source='auto', target='en')
```

```python
%%time

l = indices[r]

errors = []

for index in l:
    try:
        df.loc[index, field_translation] = translator.translate_batch([df.loc[index, field][:4500]])
    except Exception as ex:
        errors.append((index, str(ex), df.loc[index, field]))
```

```python
for index, ex, caption in errors:
    print('\033[34m', index, ex)
    print(caption)
    print('\033[0m')
```
