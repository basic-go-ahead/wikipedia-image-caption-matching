```python
import numpy as np
import pandas as pd

from deep_translator import GoogleTranslator
```

```python
images = pd.read_csv('../input/traindataset-part0-4-count-5/images_part0_between4,5.csv')

images
```

```python
translator = GoogleTranslator(source='auto', target='en')
```

```python
%%time

images['undigit_filename_translation'] = np.nan
indices = images[(images['filename_lang'] != 'en') & (~images['undigit_filename_empty'])].index

print(len(indices))

l = indices[4 * len(indices) // 5 :]
    
images.loc[l, 'undigit_filename_translation'] = translator.translate_batch(images.loc[l, 'undigit_filename'].str.slice(stop=4500).tolist())
```

```python
images.to_csv('tran_images_part0_between4,5.csv', index=False)
```
