```python
import os

import numpy as np
import pandas as pd
import cloudpickle

from transformers import XLMRobertaModel, XLMRobertaTokenizer
import torch

from sentence_transformers import SentenceTransformer, util
```

```python
df = pd.read_csv('../input/test-combiner-target-dataset/final_matchings.csv', keep_default_na=False)

df
```

### Loading Model

```python
model = SentenceTransformer('paraphrase-multilingual-mpnet-base-v2')
```

## Retrieving Embeddings

```python
caption_embeddings = model.encode(df['caption_reference_description'].values)

print(caption_embeddings.shape)

with open('caption_embeddings.data', 'wb') as file:
    cloudpickle.dump(caption_embeddings, file)
```

```python
prefinal_caption_embeddings = model.encode(df['PREfinal_caption'].values)

print(prefinal_caption_embeddings.shape)

with open('prefinal_caption_embeddings.data', 'wb') as file:
    cloudpickle.dump(prefinal_caption_embeddings, file)
```
