```python
import numpy as np
import pandas as pd

from deep_translator import GoogleTranslator
```

```python
N_PART = 1

N_PART -= 1

field = 'undigit_filename'
field_lang = 'filename_lang'
field_translation = 'filename_translation'
```

```python
df = pd.read_csv('../input/test-dataset/test-images.csv', keep_default_na=False)

df
```

```python
df[field_translation] = ''

indices = df[(df[field_lang] != 'en') & (df[field] != '') & (~df[field].str.isspace()) & (~df['undigit_filename_empty'])].index

print(len(indices))
```

```python
translator = GoogleTranslator(source='auto', target='en')
```

```python
%%time

l = indices

print(len(l))

errors = []

for index in l:
    try:
        df.loc[index, field_translation] = translator.translate_batch([df.loc[index, field][:4500]])
    except Exception as ex:
        errors.append((index, str(ex), df.loc[index, field]))
```

```python
for index, ex, caption in errors:
    print('\033[34m', index, ex)
    print(caption)
    print('\033[0m')
    print()
```
